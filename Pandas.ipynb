```python
def quicksort(arr, key = 'list'):
    slist =[]
    llist =[]
    if len(arr)<2:
        return(arr)
    else:
        pivot=arr[0]
        for i in range(1,len(arr)):
            if arr[i]<pivot:
                slist.append(arr[i])
            elif arr[i]>=pivot:
                llist.append(arr[i])              
        return quicksort(slist)+[pivot]+quicksort(llist)
```

```python
array = [1,2,32,3,32,4,43,2,3]
quicksort(array)
```

```python
from collections import deque
graph = {}
graph['you']=["Linda","Brad","Diana"]
graph["Linda"] = ["Ram"]
graph["Brad"] = ["Sam"]
graph["Diana"]= []
```

```python
search_queue = deque()
search_queue += graph['you']
def person_is_seller(name):
    return name[-1] == 'm'
```

```python
def mangos(graph):
    search_queue = deque()
    search_queue += graph['you']
    while search_queue:
        person = search_queue.popleft()
        if person_is_seller(person):
            print(person + " is a mango seller !")
            return True
        else:
            search_queue+= graph[person]
    return False 
```

```python
mangos(graph)
```
