```python
def fairRations(B):
    count = 0
    if len(B)<3:
        return 'NO'
    for i in range(1,len(B)-1):
        if B[i-1]%2 ==1 and B[i+1]%2==1 and B[i]%2 == 0:
            count+=4
            B[i]+=2
            B[i-1]+=1
            B[i+1]+=1
    if count == 0:
        return 'NO'
    else:
        return count
```
